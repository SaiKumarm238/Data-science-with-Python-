```python
values = [20,40,50]
plt.figure(figsize=(5,2.5))
plt.plot(values,"o")
plt.xticks(range(len(values)))
plt.yticks(values)
plt.title("A graph of x")
plt.xlabel("index of x")
plt.ylabel("index of y")
for i,j in enumerate(values):
    plt.text(i,j,(i,j))
```

```python
np.random.seed(10)
a = np.random.randint(10,200,50)
a
```

Creat new array <br>
a>50 and a<150 -good <br>
a<50 - bad <br>
a>150 - Excellent

```python
def number_cat(n):
    if n<50:
        return "bad"
    elif (n>50) and(n>150):
        return "good"
    else:
        return "excellent"
```

```python
number_cat(150)
```

```python
number_cat(10)
```

```python
number_cat_vect = np.vectorize(number_cat)
```

```python
number_cat_vect(a)
```

```python
import pandas as pd
cat = pd.Series(number_cat_vect(a))
cat
```

```python
cat.value_counts()
```
